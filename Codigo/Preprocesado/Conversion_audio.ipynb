```python
import pandas as pd
df=pd.read_csv('texto_final.csv')
```

```python
import os
carpeta = 'Audios_etiquetados'
archivos_en_carpeta = os.listdir(carpeta)
valores_en_df = df['video_name'].unique()
valores_en_carpeta_y_df = set(valores_en_df) & set(archivos_en_carpeta)
df_filtrado = df[df['video_name'].isin(valores_en_carpeta_y_df)]
```

```python
df_filtrado
```

```python
df_filtrado=df_filtrado.drop('text', axis=1)
```

```python
df_filtrado['video_name']=df_filtrado['video_name'].astype(str)
```

```python
import librosa
def cargar_audio(audio_path):
    audio_array, _ = librosa.load(audio_path, sr=16000)
    print(f"Audio cargado: {audio_path}, longitud: {len(audio_array)}")
    return audio_array.tolist()

df_filtrado['audio'] = None

for index, row in df_filtrado.iterrows():
    audio_path = row['video_name']
    print(f"Procesando archivo: {audio_path}")
    audio_array = cargar_audio('Audios_etiquetados/'+audio_path)
    df_filtrado.at[index, 'audio'] = audio_array
    
df_filtrado.to_csv('val_final.csv', index=False)
```
