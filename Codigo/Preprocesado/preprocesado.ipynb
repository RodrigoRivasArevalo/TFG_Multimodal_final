```python
a=0
all_frases=pd.DataFrame(columns=['ID_frase', 'text', 'speaker','emotion','video_name','conv'])
for x in df['conversation']:
    a+=1
    for y in x:
        all_frases = all_frases._append({'ID_frase': y['utterance_ID'], 'text': y['text'], 'speaker': y['speaker'],'emotion': y['emotion'],'video_name':y['video_name'],'conv':a}, ignore_index=True)
```

```python
a=0
all_emotions=pd.DataFrame(columns=['emotion', 'ID_frase','conv'])
for x in df['emotion-cause_pairs']:
    a+=1
    for y in x:
        all_emotions = all_emotions._append({'emotion': y[0], 'ID_frase': y[1],'conv':a}, ignore_index=True)
```

```python
all_frases.conv.info()
```

```python
resultado = all_emotions.merge(all_frases, left_on=['conv','ID_frase'], right_on=['conv','ID_frase'], how='inner')
print(resultado)
```

```python
all_emotions['conv'] = all_emotions['conv'].astype(int)
all_emotions['ID_frase'] = all_emotions['ID_frase'].astype(int)
all_frases['conv'] = all_frases['conv'].astype(int)
all_frases['ID_frase'] = all_frases['ID_frase'].astype(int)


df_resultado = pd.merge(all_emotions, all_frases, on=['conv', 'ID_frase'],how='left')
df_resultado
```
